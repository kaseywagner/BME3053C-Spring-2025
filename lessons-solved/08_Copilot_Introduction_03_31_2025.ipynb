# What is GitHub Copilot?

GitHub Copilot is an AI pair programmer that helps you write code faster and with less effort. It works by:

- Suggesting whole lines or blocks of code as you type
- Converting comments into working code
- Helping you understand and work with unfamiliar libraries
- Offering multiple suggestions for you to choose from
- Learning from your coding style and preferences

# Basic Usage

## 1. Code Completion

Copilot provides suggestions as you type. Let's try some examples:

1. Start typing a function definition:
```python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
```

2. Copilot will suggest the implementation. Press Tab to accept or keep typing for different suggestions.

````python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
    if height <= 0:
        raise ValueError("Height must be greater than zero.")
    if weight <= 0:
        raise ValueError("Weight must be greater than zero.")
    bmi = weight / (height ** 2)
    return round(bmi, 2)

def bmi_category(bmi):
    # Determine BMI category based on the calculated BMI
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obesity"
def main():
    try:
        weight = float(input("Enter weight in kg: "))
        height = float(input("Enter height in meters: "))
        bmi = calculate_bmi(weight, height)
        category = bmi_category(bmi)
        print(f"Your BMI is {bmi} and you are classified as {category}.")
    except ValueError as e:
        print(e)
    except Exception as e:
        print("An unexpected error occurred:", e)
if __name__ == "__main__":
    main()
````

````python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from ipywidgets import interactive
import ipywidgets as widgets

def plot_with_slope(slope):
    # Generate random data
    np.random.seed(42)  # For reproducibility
    x = np.random.rand(50) * 10
    y = slope * x + np.random.randn(50) * 2

    # Create scatter plot
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(x, y, c=y, cmap='viridis', label='Data Points')

    # Add trend line
    model = LinearRegression()
    X = x.reshape(-1, 1)
    model.fit(X, y)
    y_pred = model.predict(X)
    plt.plot(x, y_pred, color='red', label='Trend Line')

    # Set y-axis limits
    plt.ylim(-5, 5)

    # Add labels and title
    plt.xlabel('X Values')
    plt.ylabel('Y Values')
    plt.title(f'Scatter Plot with Trend Line (Slope: {slope:.2f})')

    # Add colorbar and legend
    plt.colorbar(scatter, label='Y Value Intensity')
    plt.legend()

    # Show grid
    plt.grid(True, alpha=0.3)
    plt.show()

interactive_plot = interactive(
    plot_with_slope,
    slope=widgets.FloatSlider(
        value=0.5,
        min=-5,
        max=5,
        step=0.1,
        description='Slope:',
        continuous_update=False
    )
)

display(interactive_plot)
````

````python
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt

def process_medical_image(image, sigma=2.0, normalize=True):
    """Process medical image with basic filtering and normalization."""
    if not isinstance(image, np.ndarray):
        raise TypeError("Input must be a numpy array")
    
    if image.size == 0:
        raise ValueError("Image array is empty")
        
    if not np.isfinite(image).all():
        raise ValueError("Image contains invalid values")
    
    # Create a copy to avoid modifying original
    processed = image.copy().astype(float)
    
    # Apply Gaussian filter
    processed = ndimage.gaussian_filter(processed, sigma=sigma)
    
    # Normalize if requested
    if normalize:
        min_val = np.min(processed)
        max_val = np.max(processed)
        if max_val > min_val:
            processed = (processed - min_val) / (max_val - min_val)
    
    return processed

def display_test_images():
    """Display test images before and after processing."""
    # Create a more realistic medical-like test image
    size = 100
    x, y = np.ogrid[-size:size, -size:size]
    # Create circular pattern (like a tumor or nodule)
    mask = x*x + y*y <= (size/2)**2
    test_image = np.zeros((2*size, 2*size))
    test_image[mask] = 100
    # Add noise
    test_image = test_image + np.random.normal(0, 10, test_image.shape)
    
    # Process image
    processed_image = process_medical_image(test_image, sigma=2.0, normalize=True)
    
    # Create figure
    plt.figure(figsize=(12, 6))
    
    # Original image
    plt.subplot(121)
    im1 = plt.imshow(test_image, cmap='gray')
    plt.colorbar(im1)
    plt.title('Original Image')
    
    # Processed image
    plt.subplot(122)
    im2 = plt.imshow(processed_image, cmap='gray')
    plt.colorbar(im2)
    plt.title('Processed Image')
    
    plt.tight_layout()
    plt.show()
  

# Run test
display_test_images()
````

### 截 Exercise: Create a data visualization dashboard
   - Use pandas for data manipulation
   - Create multiple types of plots using matplotlib
   - Add interactive elements

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from ipywidgets import interact, widgets

# Create sample medical dataset
np.random.seed(42)
n_patients = 100
data = pd.DataFrame({
    'Age': np.random.normal(60, 15, n_patients),
    'Blood_Pressure': np.random.normal(120, 15, n_patients),
    'Glucose': np.random.normal(100, 20, n_patients),
    'Heart_Rate': np.random.normal(75, 10, n_patients),
    'Gender': np.random.choice(['Male', 'Female'], n_patients),
    'Risk_Group': np.random.choice(['Low', 'Medium', 'High'], n_patients)
})

def plot_dashboard(plot_type='scatter'):
    """
    Create interactive medical data visualization dashboard
    
    Args:
        plot_type (str): Type of plot to display ('scatter', 'histogram', 'boxplot')
    """
    plt.figure(figsize=(12, 6))
    
    if plot_type == 'scatter':
        # Scatter plot with color coding by Risk Group
        sns.scatterplot(data=data, x='Age', y='Blood_Pressure', 
                       hue='Risk_Group', style='Gender')
        plt.title('Blood Pressure vs Age by Risk Group and Gender')
        
    elif plot_type == 'histogram':
        # Histogram of glucose levels
        sns.histplot(data=data, x='Glucose', hue='Risk_Group', 
                    multiple="stack", bins=20)
        plt.title('Distribution of Glucose Levels by Risk Group')
        
    elif plot_type == 'boxplot':
        # Box plot of heart rate by risk group
        sns.boxplot(data=data, x='Risk_Group', y='Heart_Rate')
        plt.title('Heart Rate Distribution by Risk Group')
    
    plt.tight_layout()
    plt.show()

# Create interactive widget
interact(plot_dashboard, 
        plot_type=widgets.Dropdown(
            options=['scatter', 'histogram', 'boxplot'],
            value='scatter',
            description='Plot Type:'
        ))

# Display summary statistics
print("\nSummary Statistics:")
print(data.describe())
````
